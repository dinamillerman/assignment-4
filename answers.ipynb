```python
#importing necessary libraries 
import pandas as pd
import numpy as np 

#loading in the files
envfile = "C:/Users/Dina/Downloads/assignment-4-main/assignment-4-main/data/Arrigetch_Peaks_Environmental_Data.csv"
speciesfile = "C:/Users/Dina/Downloads/assignment-4-main/assignment-4-main/data/Arrigetch_Peaks_Species_Data.csv"

#making the files data frames
env = pd.read_csv(envfile)
species = pd.read_csv(speciesfile)
```

```python
#displays first few rows of env and its dimensions
env.head()
```

```python
#displays first few rows of spcies and its dimensions
species.head()
```

```python
#replacing values that are -9999 with np.NaN
env = env.replace(-9999, np.NaN)
species = species.replace(-9999, np.NaN)
```

```python
#this prints how many unique plant community names there are
print(len(pd.unique(env.PLANT_COMMUNITY_NAME)))
```

```python
#this puts all numerical columns except turboveg_plot_number in env in a list
cols = ["ELEVATION", "ASPECT", "SLOPE", "COVER_LITTER_LAYER", "COVER_OPEN_WATER", "COVER_ROCK", "COVER_CRUST", "COVER_BARE_SOIL"]
#this prints summary statistics of all of these columns
env[cols].describe()
```

```python
#this merges the two data frames by the column TURBO_PLOT_NUMBER
merged = pd.merge(env, species, left_on="TURBOVEG_PLOT_NUMBER", right_on = "TURBOVEG_PLOT_NUMBER")
```

```python
#this prints out how many rows in the merged data frame contain missing data 
x = merged.isnull().any(axis = 1)
sum(x)
```

```python
#What species was present in the most plots?

#establishing an empty dictionary 
species_dict = {}
#for loop looping through the columns
for col in species.columns:
    #looping through all relevant columns 
    if col != "TURBOVEG_PLOT_NUMBER":
        #creating dictionary items 
        species_dict[col] = species[col][species[col]>0].shape[0]
#establishing placeholder variables 
count = 0
most_species = "placeholder"
#looping through the previously created dictionary 
for name, number in species_dict.items():
    #if the number of species is greater than the previous
    if number > count:
        #the value of the initia variables are changed
        count = number
        most_species = name
print(most_species)


```

```python
#for all rows in species, calculate the sum of all 
#the columns for each row
#add this sum as a new column called totals
totals = []
for col in species.columns:
    if col != "TURBOVEG_PLOT_NUMBER":
        col_total = sum(species[col])
        totals.append(col_total)
species["totals"] = totals #this is the issue
```

Some inconsistencies are present between the description and the data.The values are not all between 0 and 5. To correct the inconsistencies, one could remove values above 5.

```python
import seaborn as sns
#creating subset of envs with rows containing specific community names
env_newdf = env.loc[(env["PLANT_COMMUNITY_NAME"] == "Caricetum scirpoideae-rupestris") | (env["PLANT_COMMUNITY_NAME"] == "Pedicularo kanei-Caricetum glacialis") | (env["PLANT_COMMUNITY_NAME"] == "Saxifrago tricuspidatae-Artemisietum alaskanae")]
env_newdf = env_newdf.replace("Caricetum scirpoideae-rupestris", "CSR")
env_newdf = env_newdf.replace("Pedicularo kanei-Caricetum glacialis", "PKCG")
env_newdf = env_newdf.replace("Saxifrago tricuspidatae-Artemisietum alaskanae", "STAA")
sns.boxplot(data = env_newdf, y = "COVER_BARE_SOIL", x = "PLANT_COMMUNITY_NAME")

plt.show()
```

The figure tells us that there are differences in bare cover soil perventage between STAA and the other two, but not between CSR and PKCG.
